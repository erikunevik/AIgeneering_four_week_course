Use Pydantic to deserialize the json object into pydantic model

```python
data["results"][:5]
```

```python
from pydantic import BaseModel

class PokemonListResponse(BaseModel):
    count: int
    results: list[dict]
    
pokemons = PokemonListResponse.model_validate(data)
pokemons
```

```python
type(pokemons)
```

```python
pokemons.count
```

```python
pokemons.results
```

```python
from pydantic import ValidationError

class PokemonListResponse(BaseModel):
    count: str
    results: list[dict]
    
try:    
    pokemons = PokemonListResponse.model_validate(data)
except ValidationError as err:
    print(err)
```

```python
data = requests.get("https://pokeapi.co/api/v2/pokemon-species/pikachu").json()

data.keys()
```

```python
data
```

```python
for key, values in data.items():
    print(f"{key:<30} {type(values)}")
```

```python
class Pokemon(BaseModel):
    id: int
    name: str
    base_happiness: int
    capture_rate: int
    evolution_chain: dict
    gender_rate:int
    generation: dict
    
pikachu = Pokemon.model_validate(data)
pikachu
```
