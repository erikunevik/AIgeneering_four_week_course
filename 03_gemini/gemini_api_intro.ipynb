```python
from google import genai

client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain how AI works in a few words",
)

print(response.text)
```

```python

def ask_gemini(prompt, model = "gemini-2.5-flash" ):
    response = client.models.generate_content(
        model=model,
        contents=prompt,
    )
    
    return response

response = ask_gemini("Can you give me 5 facts abt Luca Toni?")
print(response.text)
```

## Thinking

```python


from google.genai import types

prompt = "Can you give me 5 facts abt Luca Toni?"

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=prompt,
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_budget=0)
    ),
)

print(response.text)
```

```python
um = response.usage_metadata

result = um.total_token_count / (um.thoughts_token_count + um.prompt_token_count)

print(result)
```

## Multimodal input

Input text and image

```python
# text_input = "Describe this image shortly"
# image_input = {"mime_type": "image/png", "data": open("bella.png", 'rb').read()}

# response = client.models.generate_content(
#     model="gemini-2.5-flash",
#     contents=dict(
#         parts=[dict(text = text_input, incline_data = image_input)]
#     )
# )

text_input = "Describe this image like she is wearing a hat"
image_input = {"mime_type": "image/png", "data": open("bella.png", 'rb').read()}


response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=dict(
        parts=[dict(text = text_input), dict(inline_data = image_input)]
    )
)

print(response.text)
```
